```python
# prompt: find duplicates

duplicate_rows = student_spending_df[student_spending_df.duplicated()]
print("Duplicate Rows :")
duplicate_rows
```

```python
#displaying shape prior to dropping duplicates
student_spending_df.shape
```

```python
# prompt: drop duplicates

# Drop duplicate rows
student_spending_df.drop_duplicates(inplace=True)
student_spending_df
```

```python
#Displaying shape after dropping duplicates
student_spending_df.shape
```

```python
# Display numeric columns
numeric_cols = student_spending_df.select_dtypes(include=[np.number]).columns
numeric_cols
```

```python
# Display non-numeric columns
cat_cols = student_spending_df.select_dtypes(exclude=[np.number]).columns
cat_cols
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = student_spending_df.isnull().sum() * 100 / len(student_spending_df)

# Round the percentages to whole numbers
missing_percentage = missing_percentage.round(0)

# Print the results with percentage signs
for col, percentage in missing_percentage.items():
  print(f"{col}: {percentage:.0f}%")
```

```python
# prompt: replace the missing values in category as 'missing' and in numeric replace by '0'

# Replace missing values in categorical columns with 'missing'
for col in cat_cols:
    student_spending_df[col] = student_spending_df[col].fillna('missing')

# Replace missing values in numerical columns with 0
for col in numeric_cols:
    student_spending_df[col] = student_spending_df[col].fillna(0)
```

```python
# Calcutes the upper and lower limits
for col in student_spending_df.select_dtypes(include=['number']).columns:
    upper_limit = student_spending_df[col].mean() + 3*student_spending_df[col].std()
    lower_limit = student_spending_df[col].mean() - 3*student_spending_df[col].std()
    print(f"Upper limit for '{col}':", upper_limit)
    print(f"Lower limit for '{col}':", lower_limit)
```

```python
#saving an imputed copy
imputed_df = student_spending_df
```

```python
# Find the outliers
imputed_df.loc[(imputed_df[col] > upper_limit) | (imputed_df[col] < lower_limit)]
```

```python
# Drop the outliers
clean_df = imputed_df.loc[(imputed_df[col] < upper_limit) & (imputed_df[col] > lower_limit)]
print('Before removing outliers:',len(imputed_df))
print('After removing outliers:',len(clean_df))
print('Outliers:', len(imputed_df)-len(clean_df))
```

## Part 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Create the bar chart
categories = ['housing', 'food', 'transportation', 'books_supplies', 'entertainment', 'personal_care', 'technology', 'health_wellness', 'miscellaneous']
avg_spending_by_category = imputed_df[categories].mean() # Calculate the mean spending for each category

sns.barplot(x=avg_spending_by_category.index, y=avg_spending_by_category.values)
plt.xlabel('Spending Category', fontsize=12)
plt.ylabel('Average Spending', fontsize=12)
plt.title('Average Student Spending by Category', fontsize=14)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Group data by gender and calculate the average spending for each category
avg_spending_by_gender = imputed_df.groupby('gender')[categories].mean()

# Transpose the DataFrame for easier plotting
avg_spending_by_gender = avg_spending_by_gender.T

# Create the bar chart
ax = avg_spending_by_gender.plot(kind='bar', figsize=(10, 6))

# Customize the chart
plt.xlabel('Spending Category', fontsize=12)
plt.ylabel('Average Spending', fontsize=12)
plt.title('Average Student Spending by Category and Gender', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left') # Move the legend outside the plot
plt.tight_layout()
plt.show()
```

```python
# Group data by year_in_school and calculate the average spending for each category
avg_spending_by_gender = imputed_df.groupby('year_in_school')[categories].mean()

# Transpose the DataFrame for easier plotting
avg_spending_by_gender = avg_spending_by_gender.T

# Create the bar chart
ax = avg_spending_by_gender.plot(kind='bar', figsize=(14, 8))

# Define bar width and positions
bar_width = 0.15
index = np.arange(len(categories))

# Customize the chart
plt.xlabel('Spending Category', fontsize=12)
plt.ylabel('Average Spending', fontsize=12)
plt.title('Average Student Spending by Category and Gender', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='year in school', bbox_to_anchor=(1.05, 1), loc='upper left') # Move the legend outside the plot
plt.tight_layout()
plt.show()
```

```python
# prompt: Pie chart of Gender in Student Spending Data and School Year in Student Spending Data

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
gender_counts = imputed_df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')

plt.subplot(1, 2, 2)
school_year_counts = imputed_df['year_in_school'].value_counts()
plt.pie(school_year_counts, labels=school_year_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('School Year Distribution')

plt.tight_layout()
plt.show()
```
